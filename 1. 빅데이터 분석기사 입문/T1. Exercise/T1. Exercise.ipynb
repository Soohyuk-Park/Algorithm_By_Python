```python
#  문제 출처 : https://www.kaggle.com/agileteam/bigdatacertificationkr/tasks?taskId=6778
#  오늘은 조금 추웠다. 그래도 돈까스가 맛있어서 기분이 좋았다. 행복을 찾고싶다는 말을 하면서 열심히 공부할거다. 나는 모자라지만 성실하고 착하니까 열심히 살거다
# mtcars dataset이 주어져있다
# 우리가 해야할 일은 qsec column을 Min_max_scale로 변환하고나서 0.5보도 큰 값을 찾으면 된다. 과정을 살펴보자.
import pandas as pd
from sklearn.preprocessing import minmax_scale # 일단 필요한 라이브러리들을 불러오자. 아직은 조금 낯설지만 익숙해져야한다.


data=pd.read_csv('../input/mtcars/mtcars.csv') #정말 간단하게 데이터를 불러오자!
#print('minmax 변환 전: \n',data['qsec'].head(10)) #여기서는 일단 스케일 변환전 어떻게 생겼는지 확인해봤다!

data['qsec'] = minmax_scale(data['qsec'])
#print(data['qsec'].head(6))

```

```python
print(sum(data['qsec'] > 0.5)) # 0.5 보다 큰 칸들이 True가 된다. 그거 다 더하면 9개.
print(len(data[data['qsec']>0.5])) # 0.5보다 큰 것들을 데이터에서 길이를 구함
#둘 다 같은 결과를 반환
```

```python
# min max를 아예 정의해서 풀 수도 있다.

def minmax(data):
    data = (data - min(data)) / (max(data) - min(data))
    return data
```
